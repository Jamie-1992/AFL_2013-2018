# Quantiles 

```python
#converting DOB to years so I can do the quantiles function 

import datetime 

AFL= pd.read_csv('stats.csv') #the dataset I used earlier, it helped having it here again for debugging. didnt index name here

def from_dob_to_age(born): # date of birth function
    today = datetime.date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
 
AFL.dtypes # D.O.B is just an object here. 

#need to convert the object 'D.O.B' to datetime first for this to work
AFL['D.O.B']= pd.to_datetime(AFL['D.O.B'])

AFL.dtypes #now a datetime object

AFL = AFL.rename(columns={'D.O.B': 'Age'}) #renaming to 'Age'

AFL['Age']= AFL['Age'].apply(from_dob_to_age) #will convert DOB to years in age 
AFL.head(20)


```

# Doing actual quantiles now after getting the D.O.B converted to years 

```python
#print(AFL['Age'].quantile([0.25,0.5,0.75,1])) #note, it shows 45 as max because dustin fletcher is included
AFL['Age'].idxmax() #finds max index which will show which player is oldest, dustin fletcher is 45 but has since retired.
AFL['Age'].idxmin()

AFL['Age'].mean() #30.08, this is higher than usual because we are considering players from 2013 who have retired now but
                  #they still get counted. 
AFL['Age'].median() # median is 30, mean and median are close which means the data is probabaly symmetric (bell curve)

AFL['Age'].var()
AFL['Age'].std() # it is 4 which suggests the data has low spread and it close to the mean 
AFL['Age'].describe() #shows all the statistics (mean, std etc.)


```

# Groupby and aggregation

```python
AFL.groupby('Position').mean() #gives the average for the categorical variable position for each column. We are 
                               #grouping by each category in position. this could answer what is the average height for each position e.g. midfield
AFL.groupby('Position')['Height'].agg([min, np.median, max]) ##grouping by position and indexing the column height only. We 
                                                             #then aggregate with min, median and max put into a list.
    
#there is a nice solution for looking at different summary statistics for different columns by creating a dictionary
#We may want mean and median for heights but min/max for weight. This is conditioned on position

AFL.groupby('Position').agg({'Height': [np.median, np.mean], 'Weight': [min, max]}) #note this can also be done like this. needs backslash to do it this way
#Note backslash allows us to continue to next line with breaking the flow.

AFL.groupby('Position')\
.agg({'Height': [np.median, np.mean], 
      'Weight': [min, max]})


```
